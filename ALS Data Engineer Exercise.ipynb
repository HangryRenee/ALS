```python
#1 people file
import numpy as np
import pandas as pd
```

```python
#import data
url = 'https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons.csv'
cons = pd.read_csv(
    url,
    sep=',' 
)
url = 'https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons_email.csv'
cons_email = pd.read_csv(
    url,
    sep=',' 
)
url = 'https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons_email_chapter_subscription.csv'
cons_sub = pd.read_csv(
    url,
    sep=','
)
```

```python
# select source , created_dt , cons_id from cons
cons_1 = pd.DataFrame(
      data = cons,
      columns=['source','create_dt','cons_id']
)
cons_1
```

```python
#select email , modified_dt , cons_id, cons_email_id from cons_email
cons_email_1 = pd.DataFrame(
      data = cons_email,
      columns=['email','modified_dt','cons_id','cons_email_id']
)
cons_email_1
```

```python
#change modified date into datetime
cons_email_1['modified_dt'] = pd.to_datetime(cons_email_1['modified_dt'])
cons_email_1.dtypes
```

```python
#merge two dfs
merge_1 = pd.merge(cons_1,cons_email_1)
merge_1
```

```python
#select is_unsub and cons_email_id from cons_sub
cons_sub_1 = pd.DataFrame(
      data = cons_sub.loc[cons_sub.chapter_id==1, :],
      columns=['isunsub','cons_email_id']
)
cons_sub_1
```

```python
people = pd.DataFrame(
      data = pd.merge(merge_1,cons_sub_1),
      columns=['email','source','isunsub','create_dt','modified_dt']
)
people
```

```python
people.rename(columns={'source':'code', 'isunsub':'is_unsub','create_dt':'created_dt','modified_dt':'updated_dt'},inplace=True)
people
```

```python
people.to_csv('people.csv', index=False)
```

```python
#to count how many unique cons_id are there for each acquisition date
#change create_dt in cons_1 from datetime(64) to object
cons_1['create_dt'] = cons_1['create_dt'].apply(lambda x: pd.Timestamp(x).strftime('%Y-%m-%d'))
cons_1.rename(columns={'create_dt':'acquisition_date','cons_id':'acquisitions'},inplace=True)
cons_1
```

```python
af = pd.DataFrame(
      data = cons_1,
      columns=['acquisition_date','acquisitions']
)
af
```

```python
#create variable "acquisitions" : number of unique cons_id for each acquisition date
acquisition_facts = af.groupby('acquisition_date')
acquisition_facts.count().reset_index()[['acquisition_date','acquisitions']].to_csv('acquisition_facts.csv', index=False)
```
